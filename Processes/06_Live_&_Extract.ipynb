# 07_Live_&_Extract..,

### Live :
> *It will be connected to a* **Live Current Data**.., *Live Connection..,*

### Extract :
> *It takes* **Snapshot of Previous Data**.., *Extract Connection..,*

> Basically used when we are unable to connect the data in Live mode..,(e.g. while travelling)
> So in that case we can pull-up the extracted data.., i.e. Static Data..,

### Data 'Refresh' Option :
### 'Live' Option :

- One Cylinder Only..,
- Live Connection..,
- Direct Refresh Option..,




### 'Extract' Option :

- Two Cylinders..,
- Extracted Connection..,
- Extract Refresh Option..,
